# Bot Hunting

```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pl.read_parquet('../data/processed/ip_features.parquet').to_pandas()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='requests_per_minute',y='unique_invoice_ids')
plt.title('RPM vs Unique Invoices')
plt.show()
```

```python
if 'success_rate' in df.columns:
    plt.figure(figsize=(10,6))
    sns.scatterplot(data=df,x='requests_per_minute',y='success_rate')
    plt.title('RPM vs Success Rate')
    plt.show()
```
