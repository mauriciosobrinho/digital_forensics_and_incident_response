# Temporal Heatmaps

```
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```
events = pl.read_parquet('../data/processed/parsed_events.parquet').to_pandas()
```

```
events['hour']=events['timestamp'].dt.hour
events['day']=events['timestamp'].dt.dayofweek
```

```
pivot=events.pivot_table(index='day',columns='hour',values='source_ip',aggfunc='count',fill_value=0)
```

```
sns.heatmap(pivot)
plt.show()
```
