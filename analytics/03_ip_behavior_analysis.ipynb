# IP Behavior Analysis

```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pl.read_parquet('../data/processed/ip_features.parquet')
```

```python
top=df.sort('total_requests',descending=True).head(20).to_pandas()
plt.figure(figsize=(12,5))
sns.barplot(data=top,x='total_requests',y='ip')
plt.title('Top 20 IPs')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['requests_per_minute'].to_list(),bins=50)
plt.title('Requests per Minute')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['unique_invoice_ids'].to_list(),bins=50)
plt.title('Unique Invoice IDs')
plt.show()
```
